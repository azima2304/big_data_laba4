# Лабораторная работа: Исследование и анализ датасетов

В ходе лабораторной работы были исследованы и визуализированы готовые датасеты. На заключительном этапе датасеты были проверены на точность. При помощи метода `train_test_split()` они были разделены на тестовую и тренировочную части. Итоговые результаты отображают процент точности прогнозов для каждого датасета:

## Результаты прогноза

| Датасет            | Точность |
|--------------------|----------|
| load_digits        | 99,11%      |
| load_wine          | 75,56%      |
| load_breast_cancer | 91,61%      |

## Анализ точности модели

### Причины низкой точности модели на датасете `load_wine`

Различия в точности модели для разных датасетов из-за особенностей данных и других факторов.

1. **Количество классов и сложность задачи**
   - **Wine**: `load_wine` представляет задачу  с тремя классами вин. Для таких алгоритмов  сложнее, так как модель должна различать несколько типов объектов.
   - **Digits и Breast Cancer**:
     - `load_digits` также многоклассовый датасет (цифры), однако изображения цифр легко классифицируются с помощью методов обработки изображений, что приводит к более высокой точности.
     - `load_breast_cancer` – бинарная классификация (наличие или отсутствие рака), что упрощает задачу классификации, так как требуется различать только два состояния. Бинарная классификация обычно достигает более высокой точности.

2. **Размеры датасетов**
   - **Wine**: В датасете `load_wine` всего 178 образцов, что может привести к переобучению модели и снижению способности обобщать результаты на новых данных.
   - **Digits и Breast Cancer**:
     - `load_digits` содержит 1797 образцов, что улучшает обобщение.
     - `load_breast_cancer` включает 569 образцов, также способствуя обобщающей способности модели.

3. **Особенности признаков (Feature Engineering)**
   - **Wine**: содержит химические характеристики вина (кислотность, магний и т.д.). Эти признаки могут пересекаться для разных классов, затрудняя их разделение.
   - **Digits и Breast Cancer**:
     - `load_digits` представляет пиксельные данные, что делает задачу подходящей для алгоритмов обработки изображений (например, K-Nearest Neighbors).
     - В `load_breast_cancer` признаки структурированы для диагностических целей, облегчая классификацию.
